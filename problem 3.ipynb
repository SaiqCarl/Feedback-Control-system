```python
import numpy as np
import control
import matplotlib.pyplot as plt
```

```python
M = 4
C = 2
K = 6
```

```python
num = [M/2, C/2, K/2]
den = [M**2/2, M*C, M*K, M*K/2, 0]
```

```python
G = control.TransferFunction(num, den)

# Perform Transient Analysis
info = control.step_info(G)
print(f"Rise Time: {info.get('RiseTime', 'N/A')}")
print(f"Settling Time: {info.get('SettlingTime', 'N/A')}")
print(f"Peak Overshoot: {info.get('PeakOvershoot', 'N/A')}")

# Create Visualization Grid
plt.figure(figsize=(10, 8))

# Plot 1: Step Response
plt.subplot(2, 2, 1)
t1, y1 = control.step_response(G)
plt.plot(t1, y1)
plt.title("Step Response")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot 2: Bode Magnitude and Phase
plt.subplot(2, 2, 2)
control.bode_plot(G, dB=True, plot=True)
plt.subplot(2, 2, 2).set_title("Bode Plot")

# Plot 3: Pole-Zero Map
plt.subplot(2, 2, 3)
control.pzmap(G)
plt.title("Pole-Zero Map")

# Plot 4: Impulse Response
plt.subplot(2, 2, 4)
t2, y2 = control.impulse_response(G)
plt.plot(t2, y2)
plt.title("Impulse Response")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
