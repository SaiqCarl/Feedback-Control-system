```python
import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
num = [1, 1]
den = [1, 4, 7, 6, 3]
```

```python
sys = control.TransferFunction(num, den)
```

```python
info = control.step_info(sys)
print(f"Rise Time: {info['RiseTime']:.4f} s")
print(f"Settling Time: {info['SettlingTime']:.4f} s")
print(f"Peak Overshoot: {info['Overshoot']:.4f} %")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
```

```python
t, y = control.step_response(sys)
axs[0, 0].plot(t, y)
axs[0, 0].set_title("Step Response")
axs[0, 0].set_xlabel("Time (s)")
axs[0, 0].set_ylabel("Amplitude")
axs[0, 0].grid(True)
```

```python
plt.sca(axs[0, 1])
control.bode_plot(sys, dB=True, plot=True)
axs[0, 1].set_title("Bode Plot")
```

```python
plt.sca(axs[1, 0])
control.pzmap(sys, plot=True)
axs[1, 0].set_title("Pole-Zero Map")
```

```python
t_imp, y_imp = control.impulse_response(sys)
axs[1, 1].plot(t_imp, y_imp)
axs[1, 1].set_title("Impulse Response")
axs[1, 1].set_xlabel("Time (s)")
axs[1, 1].set_ylabel("Amplitude")
axs[1, 1].grid(True)
```

```python
plt.tight_layout()
plt.show()
```
